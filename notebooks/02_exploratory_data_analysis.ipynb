```python
# Correlation analysis between all variables
print("=== Correlation Analysis ===")
df_sorted = df.sort_values('Year')

# Calculate correlation matrix
correlation_matrix = df_sorted[['Year', 'Deaths', 'Population', 'Age-Adjusted Rate', 'Crude Rate']].corr()

print("Correlation Matrix:")
print(correlation_matrix.round(3))

print("\nStrongest Correlations:")
# Find strongest correlations (excluding self-correlations)
correlations = []
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        if i != j:
            corr_value = correlation_matrix.iloc[i, j]
            correlations.append({
                'Variable1': correlation_matrix.columns[i],
                'Variable2': correlation_matrix.columns[j],
                'Correlation': corr_value
            })

# Sort by absolute correlation value
correlations_sorted = sorted(correlations, key=lambda x: abs(x['Correlation']), reverse=True)

# Display top correlations
for i, corr in enumerate(correlations_sorted[:5]):
    direction = "positive" if corr['Correlation'] > 0 else "negative"
    print(f"{i+1}. {corr['Variable1']} ↔ {corr['Variable2']}: {corr['Correlation']:.3f} ({direction})")

print("\nKey Insights:")
print(f"• Deaths and Population: {correlation_matrix.loc['Deaths', 'Population']:.3f}")
print(f"• Deaths and Age-Adjusted Rate: {correlation_matrix.loc['Deaths', 'Age-Adjusted Rate']:.3f}")
print(f"• Population and Age-Adjusted Rate: {correlation_matrix.loc['Population', 'Age-Adjusted Rate']:.3f}")
```

## Correlation Insights

Deaths are strongly positively correlated with both crude rate (0.92) and age-adjusted rate (0.88), which is expected given that mortality rates are derived from total deaths relative to population.

Population shows a moderate negative correlation with deaths (−0.51) and a strong negative correlation with age-adjusted rate (−0.78). This suggests that the recent increase in mortality occurred during a period of stable or declining population, reinforcing the interpretation that mortality burden increased independently of population growth.

Because the dataset spans only six years, these correlations should be interpreted descriptively rather than inferentially.

```python
# Age-Adjusted vs Crude Rate comparison
print("=== Age-Adjusted vs Crude Rate Comparison ===")
df_sorted = df.sort_values('Year')

# Calculate differences between rates
df_sorted['Rate_Difference'] = df_sorted['Age-Adjusted Rate'] - df_sorted['Crude Rate']
df_sorted['Rate_Percentage_Difference'] = ((df_sorted['Age-Adjusted Rate'] - df_sorted['Crude Rate']) / df_sorted['Crude Rate']) * 100

print("Rate Comparison by Year:")
for _, row in df_sorted.iterrows():
    print(f"{row['Year']}:")
    print(f"  Age-Adjusted: {row['Age-Adjusted Rate']:.1f}")
    print(f"  Crude Rate: {row['Crude Rate']:.1f}")
    print(f"  Difference: {row['Rate_Difference']:+.1f} ({row['Rate_Percentage_Difference']:+.1f}%)")

print("\nRate Analysis:")
print(f"• Average Age-Adjusted Rate: {df_sorted['Age-Adjusted Rate'].mean():.2f}")
print(f"• Average Crude Rate: {df_sorted['Crude Rate'].mean():.2f}")
print(f"• Average Difference: {df_sorted['Rate_Difference'].mean():.3f}")

print("\n=== Demographic vs Risk Assessment ===")
print("• Age-Adjusted Rate controls for age structure")
print("• Crude Rate reflects both age structure + risk")
print("• Difference indicates demographic impact")

print("\nRisk vs Demographic Decomposition:")
for i in range(1, len(df_sorted)):
    prev_year = df_sorted.iloc[i-1]
    curr_year = df_sorted.iloc[i]
    
    crude_change = ((curr_year['Crude Rate'] - prev_year['Crude Rate']) / prev_year['Crude Rate']) * 100
    age_adj_change = ((curr_year['Age-Adjusted Rate'] - prev_year['Age-Adjusted Rate']) / prev_year['Age-Adjusted Rate']) * 100
    demographic_impact = crude_change - age_adj_change
    risk_impact = age_adj_change
    
    print(f"{prev_year['Year']} → {curr_year['Year']}:")
    print(f"  Demographic Impact: {demographic_impact:+.1f}%")
    print(f"  True Risk Impact: {risk_impact:+.1f}%")

print("\nInterpretation:")
print("• If Age-Adjusted Rate > Crude Rate: Population aging effect")
print("• If Age-Adjusted Rate ≈ Crude Rate: True risk changes")
print("• Difference magnitude = Demographic impact strength")

print("\nCorrelation with Deaths:")
print(f"• Age-Adjusted Rate ↔ Deaths: {df_sorted['Age-Adjusted Rate'].corr(df_sorted['Deaths']):.3f}")
print(f"• Crude Rate ↔ Deaths: {df_sorted['Crude Rate'].corr(df_sorted['Deaths']):.3f}")

print("\nCorrelation with Population:")
print(f"• Age-Adjusted Rate ↔ Population: {df_sorted['Age-Adjusted Rate'].corr(df_sorted['Population']):.3f}")
print(f"• Crude Rate ↔ Population: {df_sorted['Crude Rate'].corr(df_sorted['Population']):.3f}")
```

- **Rate Differences**: Small magnitude suggests limited aging population effect

### Interpretation:

**Primary Driver:** True mortality risk changes, not demographic shifts

- Strong correlation between both metrics and death counts

Both rate metrics are reliable for this analysis, with Age-Adjusted Rate being slightly more appropriate for understanding demographic sensitivity, though differences are minimal in this dataset.
