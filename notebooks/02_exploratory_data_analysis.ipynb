```python
import pandas as pd
import numpy as np
from pathlib import Path

# Set up paths
PROJECT_ROOT = Path.cwd().parent
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"

# Load cleaned data
cleaned_file = PROCESSED_DIR / "cleaned_cancer_data.csv"
df = pd.read_csv(cleaned_file)

print(f"Loaded cleaned data: {df.shape[0]:,} rows × {df.shape[1]} columns")
print(f"Columns: {list(df.columns)}")
print("\nFirst few rows:")
print(df.head())
```

```python
# Year-over-year death trends
print("=== Year-over-Year Death Trends ===")
df_sorted = df.sort_values('Year')

for i in range(1, len(df_sorted)):
    prev_year = df_sorted.iloc[i-1]
    curr_year = df_sorted.iloc[i]
    
    deaths_change = curr_year['Deaths'] - prev_year['Deaths']
    deaths_pct_change = (deaths_change / prev_year['Deaths']) * 100
    
    print(f"{prev_year['Year']} → {curr_year['Year']}:")
    print(f"  Deaths: {prev_year['Deaths']:,} → {curr_year['Deaths']:,} ({deaths_pct_change:+.1f}%)")
    print(f"  Change: {deaths_change:+,} deaths")
```

## Year-over-Year Mortality Trends (2018–2023)

Between 2018 and 2019, total deaths declined significantly by 7.4% (−137 deaths), representing the largest single-year decrease in the observed period. Mortality remained relatively stable from 2019 to 2020, with only a slight 0.1% decrease.

Beginning in 2020, deaths increased for three consecutive years. From 2020 to 2021, deaths rose by 1.2%, followed by a 2.9% increase from 2021 to 2022, and a further 3.2% increase from 2022 to 2023. By 2023, total deaths reached 1,829 — the highest level in the dataset.

Overall, the data suggest an initial decline in mortality followed by a sustained upward trend from 2020 onward. Further analysis is needed to determine whether this increase reflects population growth, aging demographics, reporting changes, or broader public health factors.

```python
# Population growth analysis
print("=== Population Growth Analysis ===")
df_sorted = df.sort_values('Year')

for i in range(1, len(df_sorted)):
    prev_year = df_sorted.iloc[i-1]
    curr_year = df_sorted.iloc[i]
    
    pop_change = curr_year['Population'] - prev_year['Population']
    pop_pct_change = (pop_change / prev_year['Population']) * 100
    
    print(f"{prev_year['Year']} → {curr_year['Year']}:")
    print(f"  Population: {prev_year['Population']:,} → {curr_year['Population']:,}")
    print(f"  Change: {pop_change:+,} ({pop_pct_change:+.2f}%)")
```

## Population Trends and Impact on Mortality Patterns (2018–2023)

The population remained relatively stable over the study period, fluctuating within a narrow range of approximately 81.5–82.9 million. While there was a temporary increase in population between 2019 and 2020 (+1.46%), population declined in most other years, including consecutive decreases from 2020 through 2023.

Importantly, the recent rise in total deaths from 2020 onward occurred during a period of flat or declining population. For example, from 2021 to 2023, deaths increased steadily (+1.2%, +2.9%, +3.2%), while the population decreased modestly each year. This divergence suggests that the upward trend in deaths cannot be explained by population growth alone.

Because mortality increased even as the population contracted, the data indicate a genuine increase in mortality burden rather than a change driven by demographic expansion. This reinforces the importance of examining age-adjusted rates to determine whether the observed increase reflects shifts in underlying mortality risk.

```python
# Death rate analysis (deaths per 100,000 population)
print("=== Death Rate Analysis (per 100,000 population) ===")
df_sorted = df.sort_values('Year')

# Calculate death rates per 100,000 population
df_sorted['Death_Rate_per_100k'] = (df_sorted['Deaths'] / df_sorted['Population']) * 100000

print("Death Rates per 100,000 Population:")
for _, row in df_sorted.iterrows():
    print(f"  {row['Year']}: {row['Death_Rate_per_100k']:.2f} per 100,000")

print("\nYear-over-Year Death Rate Changes:")
for i in range(1, len(df_sorted)):
    prev_year = df_sorted.iloc[i-1]
    curr_year = df_sorted.iloc[i]
    
    rate_change = curr_year['Death_Rate_per_100k'] - prev_year['Death_Rate_per_100k']
    rate_pct_change = (rate_change / prev_year['Death_Rate_per_100k']) * 100
    
    print(f"{prev_year['Year']} → {curr_year['Year']}:")
    print(f"  Rate: {prev_year['Death_Rate_per_100k']:.2f} → {curr_year['Death_Rate_per_100k']:.2f} ({rate_pct_change:+.1f}%)")
    print(f"  Change: {rate_change:+.2f} per 100,000")
```

## Death Rate Analysis: Mortality Burden Independent of Population Size (2018–2023)

Death rates per 100,000 population provide a standardized measure of mortality risk that accounts for population fluctuations. This analysis reveals whether changes in total deaths reflect genuine shifts in mortality burden or simply demographic changes.

### Key Findings:

**Initial Decline (2018–2020):**
- Death rates declined from 2.25 to 2.05 per 100,000 (-9.0% total)
- Largest single-year decrease occurred 2018→2019 (-7.1%)
- Suggests improving mortality outcomes during early period

**Sustained Increase (2020–2023):**
- Death rates rose steadily for three consecutive years
- 2020→2021: +2.0%, 2021→2022: +3.3%, 2022→2023: +3.6%
- By 2023, death rates (2.24) nearly returned to 2018 peak (2.25)

### Critical Insight:

The upward mortality trend from 2020 onward is **more pronounced** when examining death rates rather than raw death counts. Death rates increased by +3.6% in 2023 compared to +3.2% in death counts, indicating that the mortality burden grew faster than population changes would suggest.

### Comparison with Population Trends:

Since death rates increased while population declined (2021–2023), the data demonstrate a **genuine increase in mortality risk** rather than demographic artifacts. This reinforces the significance of the observed trend and suggests underlying factors beyond population dynamics.

### Implications:

The convergence of death rates toward 2018 levels despite population contraction warrants further investigation into potential contributing factors such as aging demographics, changes in cancer incidence, or shifts in healthcare access and reporting practices.
