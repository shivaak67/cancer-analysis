# 2. Basic Data Exploration

Simple pandas operations to understand the cancer mortality dataset.

```python
import pandas as pd
import numpy as np
from pathlib import Path

# Set up paths
PROJECT_ROOT = Path.cwd().parent
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
RAW_DIR = PROJECT_ROOT / "data" / "raw"

print("Environment setup complete!")
```

```python
# Load data
cleaned_file = PROCESSED_DIR / "cleaned_cancer_data.csv"
raw_files = list(RAW_DIR.glob("*.csv"))

if cleaned_file.exists():
    df = pd.read_csv(cleaned_file)
    print(f"Loaded cleaned data: {df.shape[0]:,} rows × {df.shape[1]} columns")
elif raw_files:
    df = pd.read_csv(raw_files[0], low_memory=False)
    print(f"Loaded raw data: {df.shape[0]:,} rows × {df.shape[1]} columns")
else:
    raise FileNotFoundError("No data files found")
```

## Basic Data Information
